<h3> Przykład prosty (orzeł na jednolitym tle) </h3>

```python
import cv2
import matplotlib.pyplot as plt
photo = cv2.imread('orzel.jpg')
```

```python
# Otworzenie pliku z pomocą OpenCV

cv2.imshow('Okno1', photo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Przykład z matplotlib

# 1-szy parametr w liście: Współrzędne Y zdjęcia (Ustawienie parametru na ::-1 odwraca zdjęcie do góry nogami)
# 2-gi parametr w liście: Współrzędne X zdjęcia (Ustawienie parametru na ::-1 odbija zdjęcie, orzeł patrzy w lewo)
# 3-ci parametr w liście: Wartości BGR dla zdjęcia

# plt.imshow(photo[::-1,:, 0])
# plt.imshow(photo[:, :, 1])
# plt.imshow(photo[:, ::-1, 0])

# Należy pamiętać, by odwrócić ostatnią listę ze względu na to, że PLT inaczej czyta kolory niż CV2 (tj. RGB ma CV2)

plt.imshow(photo[:, :, ::-1])
```

```python
# Przekonwertowanie zdjecia do skali szarości (oba przykłady)

grayPhoto = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)
cv2.imshow('Okno1', grayPhoto)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(grayPhoto, cmap='gray')
```

```python
# Sobel koordynaty X (wyszczególnienie szczegółów zdjęcia na szarej bitmapie). Konwersja CV2, wyświetlenie plt

sobelx = cv2.Sobel(grayPhoto,cv2.CV_64F,1,0,ksize=3)
plt.imshow(sobelx, cmap='gray')
```

```python
# Sobel koordynaty X (wyszczególnienie szczegółów zdjęcia na szarej bitmapie). Konwersja CV2, wyświetlenie plt

sobely = cv2.Sobel(grayPhoto,cv2.CV_64F,0,1,ksize=3)
plt.imshow(sobely, cmap='gray')
```

```python
# Zaznaczenie (ledwo widocznych) krawędzi na zdjęciu.

laplacian = cv2.Laplacian(grayPhoto,cv2.CV_64F)
plt.imshow(laplacian, cmap="gray")
```

```python
# Wyświetlenie krawędzi, zgodnie z różnicą wartości dla zdjecia czarno - białego
edges = cv2.Canny(grayPhoto,100,200)
plt.imshow(edges, cmap="gray")
```
